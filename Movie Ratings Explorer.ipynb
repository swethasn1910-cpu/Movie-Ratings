```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import dataset

df = pd.read_csv("/Users/swethashankar/Downloads/bollywood_movie_ratings.csv")
```

```python
df.head()
```

```python
#explore data

df.info()
df.describe()
df['Type'].value_counts()
```

```python
#average rating per movie 

avg_rating = df.groupby("Title")["Rating"].mean().sort_values(ascending=False)
```

```python
#top 10 highest rated bolloywood movies

avg_rating.head(10)
```

```python
#most popular genres

df["Title"].value_counts().head(10)
```

```python
#Distribute of ratings with histplot

sns.histplot(df["Rating"], bins=5)
plt.title("Distribution of the bollywood movie ratings")
plt.show()
```

```python
#Rating by year

df.groupby("Year")["Rating"].mean().plot(kind="line")
plt.title("Average bollywood movie rsting by year")
plt.show()
```

```python
# Calculate average rating per movie

avg_rating = df.groupby("Title")["Rating"].mean().sort_values(ascending=False).head(10)
```

```python
#top 10 bollywood movies average rating
sns.barplot(x=avg_rating.values, y=avg_rating.index)
plt.title("average of top 10 bollywood movies")
plt.xlabel("Average Ratings")
plt.ylabel("Movie Title")
plt.show()
```

```python
#genre distribution via pie chart

type_counts = df["Type"].value_counts() #to count the frequency of each movie 
```

```python
plt.figure(figsize=(8,8))
plt.pie(type_counts, labels=type_counts.index, autopct='&1.1f%%', startangle=140)
plt.title("Distribution of bollywood movies types")
plt.show()
```
